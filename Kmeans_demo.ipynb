## Demo of my kmeans implementation

### Import the necessary modules.

```python
from Kmeans_from_scratch import Kmeans
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [20, 10]
```

### Instantiate the Kmeans class with 3 clusters and for 15 iterations.

```python
kmeans = Kmeans(n_clusters = 3, max_iterations = 15)
```

### Load an image for clustering

```python
data = plt.imread('jeep.jpg')
plt.title('Original image', fontsize = 20)
plt.imshow(data)
plt.show()
```

### Fit the centroids of the Kmeans using the image data.

```python
centroids, labels = kmeans.fit(data)
```

### Plot the centroid pixel values for dispalay purposes.

```python
print(centroids)

for i in range(centroids.shape[0]):
    show_test = np.full((50, 50, 3), centroids[i])
    plt.subplot(1, centroids.shape[0], i+1)
    plt.title('Centroid {}'.format(i), fontsize = 20)
    plt.imshow(show_test)
plt.show()
```

### Predict on the image data with the fitted centroids.

```python
res = kmeans.predict(data)
```

### Plot the result clustered image.

```python
plt.imshow(res)
plt.title('Clustered image', fontsize = 20)
plt.show()
```
